```python
import numpy as np
import pandas as pd

vehicles = pd.read_csv('vehicles.csv')
```

```python
vehicles.head(10)
```

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

```python
vehicles.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
correlation = vehicles.corr()
fig, ax = plt.subplots(figsize=(20, 14))
sns.heatmap(correlation, vmax=1, annot=True, linewidths=.5, ax=ax)
```

```python
plt.scatter(vehicles['City MPG'], vehicles['CO2 Emission Grams/Mile'])
plt.title("Relationship")
plt.xlabel("City MPG")
plt.ylabel("CO2 Emission Grams/Mile")
plt.show()
```

```python
fig = sns.boxplot(vehicles['Year'], vehicles['Combined MPG'], data=vehicles)

fig.axis(ymin=0, ymax=vehicles['Combined MPG'].max())
```
