### breakdown of students who participate in extracurricular activities by sex

```python
student.groupby(['sex', 'activities']).agg({'activities':'count'})
```

```python
pd.crosstab(index=student.sex, columns=student.activities)
```

### type of address vs. family size:

```python
student.groupby(['address', 'famsize']).agg({'famsize':'count'})
```

```python
pd.crosstab(index=student.address, columns=student.famsize)
```

### how many students improved their grade between the first and second period

```python
student['improvement'] = np.where(student.G2 > student.G1, "improved", "did not improve")
```

```python
student.groupby(['school', 'improvement']).agg({'improvement':'count'})
```

```python
pd.crosstab(index=student.school, columns=student.improvement)
```
