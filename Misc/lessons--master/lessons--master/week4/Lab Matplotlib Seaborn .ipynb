```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
sns.lineplot(x,y)
sns.lineplot(x,z)
```

```python
fig, [ax1, ax2] = plt.subplots(1,2)
ax1 = sns.lineplot(x,y,ax=ax1)
ax2 = sns.lineplot(x,z,ax=ax2)
```
