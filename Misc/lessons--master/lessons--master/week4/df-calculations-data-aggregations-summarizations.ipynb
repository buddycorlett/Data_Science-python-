```python
import numpy as np
import pandas as pd

vehicles = pd.read_csv('vehicles.csv')
```

```python
vehicles.head(10)
```

```python
vehicles.groupby(['Transmission'])['Highway MPG', 'City MPG', 'Combined MPG'].mean()
```

```python
vehicles.groupby(['Fuel Type', 'Cylinders','Drivetrain'])['CO2 Emission Grams/Mile'].median()
```

```python
vehicles.groupby(['Fuel Type','Cylinders']).agg({'Combined MPG':['mean', 'median', 'std']})
```

```python
vehicles.groupby(['Fuel Type','Cylinders'])['Combined MPG'].agg(['mean', 'median', 'std'])
```

```python
vehicles.groupby(['Fuel Type']).agg({'Combined MPG':['mean', 'median', 'std']})
```

```python
from scipy import stats

def agg_mode(x):
    return(stats.mode(x)[0])
```

```python
stats.mode(vehicles['Vehicle Class'])[0][0]
```

```python
vehicles.groupby("Transmission")["Vehicle Class"].agg(agg_mode)
```
