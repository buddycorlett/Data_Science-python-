```python
from scipy.interpolate import interp1d
x = np.arange(0, 11, 0.1)
y = np.cos(x)
plt.plot(x,y)
```

```python
f = interp1d(x, y)
xnew = np.arange(0, 5.25, 1)
plt.plot(xnew, f(xnew))
```

```python
f = interp1d(x, y, 'quadratic')
xnew = np.arange(0, 10.25, 0.25)
plt.plot(xnew, f(xnew))
```
