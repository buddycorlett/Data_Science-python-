```python
import pandas as pd
import numpy as np
vehicles = pd.read_csv('vehicles.csv')
```

```python
vehicles.head(5)
```

```python
print(vehicles['Vehicle Class'].unique())
len(vehicles['Vehicle Class'].unique())
```

```python
vehicles.pivot_table(index=["Vehicle Class"])
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"])
```

```python
vehicles['Cylinders'].unique()
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"]).shape
```

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG"])
```

```python
vehicles.groupby('Vehicle Class').agg({'Combined MPG':'mean'})
```

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG"], aggfunc=np.sum)
```

```python
vehicles.groupby('Vehicle Class').agg({'Combined MPG':'sum'})
```

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG", "CO2 Emission Grams/Mile"])
```

```python
vehicles.groupby('Vehicle Class').agg({'Combined MPG':'mean', 'CO2 Emission Grams/Mile':'mean'})
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=["Combined MPG"])
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=["Combined MPG"], fill_value=0)
```
