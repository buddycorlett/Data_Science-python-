### Basic Plots

Line Plots

```python
x1 = np.arange(0,10,1)
print(x1)
len(x1)
```

```python
x2 = np.linspace(0,5,10)
print(x2)
```

```python
plt.plot(x1, x2, label='linear')
plt.plot(x1, x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Linear vs Non Linear Plots")
plt.legend()

plt.show()
```

QUICK QUIZ 

PLot a cosine curve with labels and title

```python
x = np.arange(1,10,.1)
y = np.cos(x)
plt.plot(x,y, label = "cosine curve")
plt.xlabel('x')
plt.ylabel('cosine')
plt.show()
```

Making subplots 

```python
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x1,x2)
ax1.set_title('LINEAR PLOT')   #Note how we are using set_ for subplots
ax1.set_xlabel('x label')      #Note how we are using set_ for subplots
ax1.set_ylabel('y label')      #Note how we are using set_ for subplots 

ax2.plot(x,y)
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

```python
def my_plots(ax, x,y, params):
    return(ax.plot(x,y,**params))

fig, (ax1, ax2) = plt.subplots(1, 2)
my_plots(ax1, x1, x2, {'marker': 'x'})
my_plots(ax2, x, y, {'marker': 'o'})

ax1.set_title('LINEAR PLOT')
ax2.set_title('COSINE PLOT')
```

Scatter Plots

```python
x = np.random.randn(10)
y = np.random.randn(10)
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

Pie Charts

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 45, 15]
explode = (0, 0.05, 0)  # only "explode" the 2nd slice (i.e. 'Non-Vegetarians')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
# ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
